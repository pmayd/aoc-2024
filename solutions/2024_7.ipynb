```python
def solution_a(data: str) -> tuple[list[str], int]:
    # Operators are always evaluated left-to-right, 
    # not according to precedence rules. 
    # Furthermore, numbers in the equations cannot be rearranged.
    # There are two different types of operators: add (+) and multiply (*).
    correct_equations = []
    final_sum = 0
    operators = ['+', '*']
    
    for equation in data.splitlines():
        #print(equation)
        left_side, right_side = equation.split(':')
        left_side = int(left_side)
        numbers = [int(x) for x in right_side.split()]
        num_operators = len(numbers) - 1
        possible_operators = list(product(operators, repeat=num_operators))
        
        for operator_order in possible_operators:
            #print(operator_order)
            sum_ = numbers[0]
            for i, operator in enumerate(operator_order):
                sum_ = eval(f"{sum_} {operator} {numbers[i+1]}")
                #print(sum_)
                
            if sum_ == left_side:
                print(f"{equation} is correct for operators {operator_order}")
                correct_equations.append(equation)
                final_sum += sum_
                break
        
    return correct_equations, final_sum
```

```python
assert solution_a(puzzle.examples[0].input_data)[1] == int(puzzle.examples[0].answer_a)
```

```python
def solution_b(data: str) -> tuple[list[str], int]:
    # Operators are always evaluated left-to-right, 
    # not according to precedence rules. 
    # Furthermore, numbers in the equations cannot be rearranged.
    # There are three different types of operators: add (+), multiply (*) and concatenate (||).
    correct_equations = []
    final_sum = 0
    operators = ['+', '*', '||']
    
    for equation in tqdm(data.splitlines(), desc="Solving equation..."):
        #print(equation)
        left_side, right_side = equation.split(':')
        left_side = int(left_side)
        numbers = [int(x) for x in right_side.split()]
        num_operators = len(numbers) - 1
        possible_operators = list(product(operators[:2], repeat=num_operators))
        
        for operator_order in possible_operators:
            #print(operator_order)
            sum_ = numbers[0]
            for i, operator in enumerate(operator_order):
                if operator == '||':
                    sum_ = eval(f"{sum_}{numbers[i+1]}")
                else:
                    sum_ = eval(f"{sum_} {operator} {numbers[i+1]}")
                #print(sum_)
                
            if sum_ == left_side:
                print(f"{equation} is correct for operators {operator_order}")
                correct_equations.append(equation)
                final_sum += sum_
                break
        
        # if the equation is not correct with + and *, try with ||
        if sum_ != left_side:
            possible_operators = list(product(operators, repeat=num_operators))
            possible_operators = [x for x in possible_operators if '||' in x]
            
            for operator_order in possible_operators:
                #print(operator_order)
                sum_ = numbers[0]
                for i, operator in enumerate(operator_order):
                    if operator == '||':
                        sum_ = eval(f"{sum_}{numbers[i+1]}")
                    else:
                        sum_ = eval(f"{sum_} {operator} {numbers[i+1]}")
                    #print(sum_)
                    
                if sum_ == left_side:
                    print(f"{equation} is correct for operators {operator_order}")
                    correct_equations.append(equation)
                    final_sum += sum_
                    break
        
    return correct_equations, final_sum
```
